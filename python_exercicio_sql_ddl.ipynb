#Mini Mundo: Sistema de Gestão Veterinária

SISTEMA VETERINÁRIO
Descrição Geral
Uma clínica veterinária deseja informatizar seu processo de gestão, incluindo o cadastro de animais, donos, consultas, tratamentos e medicamentos. O sistema deve permitir um controle eficiente das operações, garantindo agilidade e organização.

# Entidades e Relacionamentos
    1. Dono
        ◦ Cada dono pode ter vários animais.
        ◦ Atributos:
            ▪ ID_Dono (PK)
            ▪ Nome
            ▪ CPF
            ▪ Telefone
            ▪ Endereço
    2. Animal
        ◦ Cada animal pertence a um único dono.
        ◦ Atributos:
            ▪ ID_Animal (PK)
            ▪ Nome
            ▪ Espécie
            ▪ Raça
            ▪ Data_Nascimento
            ▪ Sexo
            ▪ ID_Dono (FK)
    3. Consulta
        ◦ Cada consulta é realizada para um único animal e atendida por um veterinário.
        ◦ Atributos:
            ▪ ID_Consulta (PK)
            ▪ Data
            ▪ Hora
            ▪ Observações
            ▪ ID_Animal (FK)
            ▪ ID_Veterinario (FK)
    4. Veterinário
        ◦ Cada veterinário pode atender várias consultas.
        ◦ Atributos:
            ▪ ID_Veterinario (PK)
            ▪ Nome
            ▪ CRMV
            ▪ Especialidade
            ▪ Telefone
    5. Tratamento
        ◦ Cada tratamento é prescrito em uma consulta e pode envolver vários medicamentos.
        ◦ Atributos:
            ▪ ID_Tratamento (PK)
            ▪ Descrição
            ▪ Data_Inicio
            ▪ Data_Fim
            ▪ ID_Consulta (FK)
    6. Medicamento
        ◦ Cada medicamento pode ser utilizado em vários tratamentos.
        ◦ Atributos:
            ▪ ID_Medicamento (PK)
            ▪ Nome
            ▪ Fabricante
            ▪ Dosagem
    7. Tratamento_Medicamento
        ◦ Entidade de relacionamento entre Tratamento e Medicamento.
        ◦ Atributos:
            ▪ ID_Tratamento (FK)
            ▪ ID_Medicamento (FK)
            ▪ Quantidade
            ▪ Frequência
    1. Pagamento (Relacionada a Consultas ou Tratamentos)
        ◦ Motivo: Consultas geralmente geram pagamentos.
        ◦ Atributos:
            ▪ ID_Pagamento (PK)
            ▪ Valor
            ▪ Data_Pagamento
            ▪ Método_Pagamento
            ▪ ID_Consulta (FK)
    2. Exame (Realizado em Animais, relacionado às Consultas)
        ◦ Motivo: Consultas podem exigir exames.
        ◦ Atributos:
            ▪ ID_Exame (PK)
            ▪ Tipo_Exame
            ▪ Resultado
            ▪ Data_Exame
            ▪ ID_Consulta (FK)
    3. Agendamento (Antes da Consulta, para planejamento)
        ◦ Motivo: Organizar consultas previamente.
        ◦ Atributos:
            ▪ ID_Agendamento (PK)
            ▪ Data_Agendamento
            ▪ Hora_Agendamento
            ▪ Status
            ▪ ID_Animal (FK)
            ▪ ID_Veterinario (FK)

```python
def consulta(query, tabela):
  connection = mysql.connector.connect(
      host = host,
      user = user,
      password = password,
      database = database
  )
  cursor = connection.cursor()

  try:
    cursor.execute(query)
    result = cursor.fetchall()
    globals()[tabela] = pd.DataFrame(result, columns=cursor.column_names)
    display(globals()[tabela])
  finally:
    cursor.close()
    connection.close()
```

```python
def ddl(comando):
    connection = mysql.connector.connect(
        host=host,
        user=user,
        password=password,
        database=database
    )
    cursor = connection.cursor()

    try:
        cursor.execute(comando)
        connection.commit()  # Confirma as mudanças no banco
        print(f"Comando DDL executado com sucesso:\n{comando}")
    except mysql.connector.Error as err:
        print(f"Erro ao executar o comando DDL: {err}")
    finally:
        cursor.close()
        connection.close()
```
